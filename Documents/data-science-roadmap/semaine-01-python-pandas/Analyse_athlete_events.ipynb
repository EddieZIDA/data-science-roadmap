```python
#Création d'un nouveau dataframe à partir des colonnes sex,age,sport,city
data=df[['Sex','Age','Sport','City']]
data
```

```python
#Remplacement des caractères F,M par des chiffres 0,1 dans la colonne sex
data['Sex'].map({'F': 0, 'M': 1})
```

```python
#Affichage des personnes avec un age compris entre 30 et 33ans
data[(data['Age']>=30) & (data['Age']<33)]
```

```python
#Age moyen des athlètes par sport
df.groupby("Sport")["Age"].mean().sort_values(ascending=False).head(10)
```

```python
#Les équipes avec le plus de médailles d'or
df[df["Medal"] == "Gold"]["Team"].value_counts().head(10)
```

```python
#Evolution du nombre de femmes participantes par année
df_femmes = df[df["Sex"] == "F"]
df_femmes.groupby("Year")["Name"].nunique().plot(kind="line", figsize=(10,5), marker="o")
plt.title("Évolution du nombre de femmes participantes")
plt.xlabel("Année")
plt.ylabel("Nombre d'athlètes")
plt.grid(True)
plt.show()
```
